```python
from scipy.stats import binom
n = 10  # number of items in the sample
p = 0.05  # probability of a defective item
q = 1 - p  # probability of a non-defective item

# 2.A. Probability that none of the items are defective?
probability_none_defective = binom.pmf(0, n, p)
print("Probability that none of the items are defective:", probability_none_defective)

# 2.B. Probability that exactly one of the items is defective?
probability_one_defective = binom.pmf(1, n, p)
print("Probability that exactly one of the items is defective:", probability_one_defective)

# 2.C. Probability that two or fewer of the items are defective?
probability_two_or_fewer_defective = binom.cdf(2, n, p)
print("Probability that two or fewer of the items are defective:", probability_two_or_fewer_defective)

# 2.D. Probability that three or more of the items are defective?
probability_three_or_more_defective = 1 - binom.cdf(2, n, p)
print("Probability that three or more of the items are defective:", probability_three_or_more_defective)
```

```python
from scipy.stats import poisson

average_cars_sold_per_week = 3

# 3.A. Probability that in a given week he will sell some cars?
probability_sell_some_cars = 1 - poisson.pmf(0, average_cars_sold_per_week)
print("Probability that in a given week he will sell some cars:", probability_sell_some_cars)

# 3.B. Probability that in a given week he will sell 2 or more but less than 5 cars?
probability_sell_2_to_4_cars = np.sum(poisson.pmf(np.arange(2, 5), average_cars_sold_per_week))
print("Probability that in a given week he will sell 2 or more but less than 5 cars:", probability_sell_2_to_4_cars)

# 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
x = np.arange(0, 10) 
cumulative_probability = poisson.cdf(x, average_cars_sold_per_week)

plt.plot(x, cumulative_probability, 'bo-', label='Poisson Distribution')
plt.title('Poisson Distribution for Cars Sold Per Week')
plt.xlabel('Number of Cars Sold Per Week')
plt.ylabel('Cumulative Probability')
plt.legend()
```

```python
#D. Check the number of outliers greater than the upper fence. [1 Mark]

Q1 = cu['Funds_in_million'].quantile(0.25)
Q3 = cu['Funds_in_million'].quantile(0.75)
IQR = Q3 - Q1

upper_fence = Q3 + 1.5 * IQR

outliers_count = (cu['Funds_in_million'] > upper_fence).sum()

print(f"Number of outliers greater than the upper fence: {outliers_count}")
```

```python
##Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down? [1 Mark]
from scipy.stats import ttest_ind

# Create datasets for companies still operating and closed down
operating_companies = cu[cu['OperatingState'] == 'Operating']['Funds_in_million'].dropna()
closed_companies = cu[cu['OperatingState'] == 'Closed']['Funds_in_million'].dropna()

# Perform independent samples t-test
t_stat, p_value = ttest_ind(operating_companies, closed_companies, equal_var=True)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret the results
if p_value < 0.05:
    print("There is a significant difference in funds raised between companies still operating and closed down.")
else:
    print("There is no significant difference in funds raised between companies still operating and closed down.")
```

#Write the null hypothesis and alternative hypothesis
Ans - Null Hypothesis : There is no significant difference in the mean funds raised between companies that are still operating and companies that have closed down.
Alternative Hypothesis : There is a significant difference in the mean funds raised between companies that are still operating and companies that have closed down.

```python
#Test for significance and conclusion [1 Mark]

alpha = 0.05  # Significance level

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Make a conclusion
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant difference in funds raised between companies still operating and closed down.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in funds raised between the two groups.")
```

```python
#Write your hypothesis comparing the proportion of companies that are operating between winners and contestants: 
import statsmodels.api as sm

# Calculate the number of winners still operating
winners_operating = cu[(cu['Result'] == 'Winner') & (cu['OperatingState'] == 'Operating')].shape[0]

# Calculate the number of contestants still operating
contestants_operating = cu[cu['OperatingState'] == 'Operating'].shape[0]

# Calculate the total number of winners
total_winners = cu[cu['Result'] == 'Winner'].shape[0]

# Calculate the total number of contestants
total_contestants = cu.shape[0]

# Perform the Z-test for proportions
z_stat, p_value = sm.stats.proportions_ztest([winners_operating, contestants_operating], [total_winners, total_contestants])

print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Make a conclusion
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant difference in the proportion of companies still operating between winners and contestants.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the proportion of companies still operating between the two groups.")
```
